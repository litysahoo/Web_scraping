```python
final= pd.DataFrame()

for k in range(1,11):
    url="https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}".format(k)
    headers = {  "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
    webpage= requests.get(url, headers=headers).text
    soup=BeautifulSoup(webpage, "lxml")
    company=soup.find_all("div", class_="companyCardWrapper")
    name=[]
    rating=[]
    ctype=[]
    ctypenew=[]
    employees=[]
    sector=[]
    old=[]
    location=[]
    m=0
    for i in company:
        ctype.append(i.find("span", class_="companyCardWrapper__interLinking").text.strip())
        if ctype[m].count(" | ")==4:
            name.append(i.find("h2").text.strip())
            rating.append(i.find("span", class_="companyCardWrapper__companyRatingValue").text.strip())
            m=m+1
        else:
            m=m+1
    for j in ctype:
        l=j.split(" | ")
        
        if len(l)>4:
            ctypenew.append(l[0])
            employees.append(l[1])
            sector.append(l[2])
            old.append(l[3])
            location.append(l[4])
        else:
            continue
    
    df_new_row = pd.DataFrame({ 'Name_of_Company': name, 'Ratings': rating, 'Type_of_Company':ctypenew, 'Sector':sector, 'Old':old, 'Location':location, 'Number_of_Employees':employees })
    final = pd.concat([final, df_new_row], ignore_index=True)

final
```
